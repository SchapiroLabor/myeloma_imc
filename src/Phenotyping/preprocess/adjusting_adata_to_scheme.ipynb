```python
%%capture
base_dir = "/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/phenotyping/uncorrected/standard"

# Define the path for the phenotype CSV file
phenotype_csv_path = '/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/src/Phenotyping/scimap/phenotyping_scheme_scimap.csv'
phenotype = pd.read_csv(phenotype_csv_path)

# Loop through each file in the directory
for file_name in os.listdir(base_dir):
    if file_name.endswith(".h5ad"):
        # Construct the full file path
        file_path = os.path.join(base_dir, file_name)
        file = file_name.split("denoised")[0]
        # Read the .h5ad file
        adata = sc.read_h5ad(file_path)
        manual_gate = pd.read_csv(f'/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/phenotyping/uncorrected/standard/{file}_manual_gates_scimap_denoised.csv')
       
        adata = sm.pp.rescale (adata, gate=manual_gate)
        # Phenotype the cells
        adata = sm.tl.phenotype_cells(adata, phenotype=phenotype, label="phenotype")

        # Save the modified .h5ad file back to the same location
        adata.write(file_path)
```

# Adjusting adatas to batch corrected values

```python
base_dir = "/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/phenotyping/uncorrected/standard"
csv_dir = '/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/standard/batch_corrected_scanorama'
save_dir = '/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/phenotyping/scanorama_corrected/standard'


# Loop through each file in the directory
for file_name in os.listdir(base_dir):
    if file_name.endswith(".h5ad"):
        print(f'processing file {file_name}')
        # Construct the full file path
        file_path = os.path.join(base_dir, file_name)

        # Read the .h5ad file
        adata = sc.read_h5ad(file_path)
        batch = pd.read_csv(os.path.join(csv_dir, file_name.split('denoised')[0] + '.csv'))
        batch.drop(columns=['CD98'], inplace=True)
        batch.rename(columns={'Object': 'CellID'}, inplace=True)
        adata = adata[adata.obs['CellID'].isin(batch['CellID'])]
        batch = batch[batch['CellID'].isin(adata.obs['CellID'])]
        #assert set(batch.columns[:-1]) == set(adata.var_names), "Gene/feature names in DataFrame and AnnData do not match"
        batch = batch.reindex(columns=adata.var_names.tolist() + ['CellID'])
        adata.X = batch.iloc[:, :-1].values


        # Save the modified .h5ad file back to the same location
        adata.write(os.path.join(save_dir, file_name))
```
