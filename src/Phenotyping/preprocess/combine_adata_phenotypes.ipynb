```python
import anndata as ad
import numpy as np
import pandas as pd
import os
import glob
```

```python
processed = "standard"
```

```python
adata_dir = f'/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/github/myeloma_standal/phenotyping/uncorrected/{processed}/'
csv_dir = f'/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/{processed}/merged_csv/'
batch_dir= f'/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/{processed}/batch_corrected_scanorama/'
adata_files = glob.glob(os.path.join(adata_dir, '*.h5ad'))
```

```python
all_dataframes = [] 

for adata_file in adata_files:
    unique_id = os.path.basename(adata_file).split(f'{processed}')[0]
    
    csv_file = glob.glob(os.path.join(csv_dir, f'*{unique_id}*.csv'))
    if not csv_file:
        print(f"No CSV file found for {unique_id}")
        continue  
    csv_file = csv_file[0]  
    
    adata = ad.read_h5ad(adata_file)
    csv = pd.read_csv(csv_file)
    csv = csv[csv['area'] >= 4]
    print(f'this is {unique_id}')
    print(adata.obs.phenotype.isna().sum())
    print(adata.obs.CellID.isna().sum())

    phenotype_map = adata.obs.set_index('CellID')['phenotype'].to_dict()
    imageid_map = adata.obs.set_index('CellID')['imageid'].to_dict()
    csv['phenotype'] = csv['Object'].map(phenotype_map)
    csv['imageid'] = csv['Object'].map(imageid_map)

    csv = csv.dropna(subset=['phenotype', 'imageid'])
    
    csv.drop(columns=["HistoneH3", "CD98", " 1", " 2", " 3", " 4", " 5", " 6", "191Ir", "193Ir"], inplace=True)
    

    csv.rename(columns={'centroid-0': 'Y_centroid', 'centroid-1': 'X_centroid', 'Object': 'CellID'}, inplace=True)
    csv['index'] = csv['imageid'] + '_' + csv['CellID'].astype(str)
    csv.set_index('index', inplace=True)
    csv.drop(columns=['CellID', 'imageid'], inplace=True)

    all_dataframes.append(csv)

final_csv = pd.concat(all_dataframes, ignore_index=False)
```

```python
# Using batch corrected values for phenotyping
all_dataframes = [] 

for adata_file in adata_files:
    unique_id = os.path.basename(adata_file).split(f'{processed}')[0]
    
    batch_file = glob.glob(os.path.join(batch_dir, f'*{unique_id}*.csv'))
    if not batch_file:
            print(f"No batch CSV file found for {unique_id}")
            continue  
    adata = ad.read_h5ad(adata_file)

    batch = pd.read_csv(batch_file[0])
    batch = batch[batch['area'] >= 4]
    batch.rename(columns={'Object': 'CellID'}, inplace=True)

    print(f'this is {unique_id}')
    print(adata.obs.phenotype.isna().sum())
    print(adata.obs.CellID.isna().sum())

    phenotype_map = adata.obs.set_index('CellID')['phenotype'].to_dict()
    imageid_map = adata.obs.set_index('CellID')['imageid'].to_dict()
    batch['phenotype'] = batch['CellID'].map(phenotype_map)
    batch['imageid'] = batch['CellID'].map(imageid_map)

    print(batch['phenotype'].isna().sum())
    batch = batch.dropna(subset=['phenotype', 'imageid'])
    
    batch.drop(columns=["HistoneH3", "CD98", " 1", " 2", " 3", " 4", " 5", " 6", "191Ir", "193Ir"], inplace=True)
    

    batch.rename(columns={'centroid-0': 'Y_centroid', 'centroid-1': 'X_centroid', 'Object': 'CellID'}, inplace=True)
    batch['index'] = batch['imageid'] + '_' + batch['CellID'].astype(str)
    batch.set_index('index', inplace=True)
    batch.drop(columns=['CellID', 'imageid'], inplace=True)

    all_dataframes.append(batch)

final_csv = pd.concat(all_dataframes, ignore_index=False)
```
