```python
import pandas as pd
import napari
from skimage.io import imread
import os
import glob
```

```python
df = pd.DataFrame(columns=['image_file', 'coordinates'])
directory = '/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/standard/img'
```

```python
file_names = ["IMC01_UB_001","IMC03_MGUS_001","IMC03_MGUS_002","IMC08_UB_002","IMC23_B_001","IMC33_B_002", "IMC34_B_001","IMC49_B_002","IMC58_B_002","IMC64_B_002","IMC86_B_002"]
```

```python
def load_image_and_annotate(file_path):
    image = imread(file_path)
    image_name = os.path.basename(file_path)

    # Create a Napari viewer
    viewer = napari.Viewer(title=f"Annotating: {image_name}")
    viewer.add_image(image, name=image_name)
    shapes_layer = viewer.add_shapes(name='Annotations')

    # Block until the viewer is closed
    viewer.show(block=True)

    shapes_data = shapes_layer.data

    # Close the viewer
    viewer.close()

    return shapes_data
```

```python
for file_name in file_names:
    file_pattern = os.path.join(directory, '*' + file_name + '*')
    file_list = glob.glob(file_pattern)
    if file_list:
        for file_path in file_list:
            coordinates = load_image_and_annotate(file_path)

            temp_df = pd.DataFrame({
    'image_file': [os.path.basename(file_path)],
    'coordinates': [str([tuple(coord.tolist()) for coord in coordinates])]
})

            df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"No matching file found for {file_name}")
```
